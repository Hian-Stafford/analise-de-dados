# 1. Trazer a base de dados para o python e ver o que tem

Importa a biblioteca pandas para ajudar na análise de dados

```python
import pandas as pd #importei a biblioteca pandas pra ajudar na análise de dados

tab_vendas = pd.read_excel('Vendas.xlsx') #lê a tabela excel chamada "Vendas" e guarda na variável "tab_vendas"
tab_vendas.head(50)
#display(tab_vendas) #mostra na tela a tabela completa
```

# 2. Pegar um panorama geral sobre a base de dados

Calcula o faturamento total 

```python
fat_total = tab_vendas['Valor Final'].sum()
print(fat_total)
```

# 3. Começar a análise de top -> down

Faturamento por loja

```python
fat_loja = tab_vendas[["ID Loja", "Valor Final"]].groupby("ID Loja").sum()
display(fat_loja)
```

# 4. Entrar nos detalhes para entender

Faturamento por loja

```python
fat_prod = tab_vendas[['ID Loja', 'Produto', 'Valor Final']].groupby(['ID Loja', 'Produto']).sum()
fat_prod.head(50)
```

```python
#tab_vendas = tab_vendas[['ID Loja']].groupby('ID Loja').value_counts()

total_nome_loja = tab_vendas.loc[(tab_vendas['ID Loja'] == 'Ribeirão Shopping')].value_counts().sum()

print(f'Ribeirão Shopping: {total_nome_loja}')
```
